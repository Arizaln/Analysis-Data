# import library

```python
import pandas as pd 
import matplotlib.pyplot as plt 
import numpy as np
pd.set_option('display.max_columns',None)
```

# data preprocesing

```python
df=pd.read_csv('/content/train.csv')
df.head(20)
```

```python
df.info()
```

```python
obj = df.isnull().sum()
for key,value in obj.iteritems():
    print(key,",",value)
```

```python
df=df.drop(['Alley','PoolQC','Fence','MiscFeature','FireplaceQu','Id'],axis=1)
```

```python
df.head(10)
```

```python
df.shape
```

```python
obj = df.isnull().sum()
for key,value in obj.iteritems():
    print(key,",",value)
```

```python
df.dropna(inplace=True)
```

```python
df.head(10)
```

```python
obj = df.isnull().sum()
for key,value in obj.iteritems():
    print(key,",",value)
```

```python
df1=pd.get_dummies(df)
```

```python
df1.head(10)
```

# define x and y

```python
x = df1.drop(['SalePrice'],axis=1).values
y = df1['SalePrice'].values
```

```python
# cek value properly added to x and y
print(x)
```

```python
print(y)
```

# split the dataset in training set and test set

```python
# to split the data set we are using the scikit library
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3, random_state=0)
```

# train the model on the training set

```python
from sklearn.linear_model import LinearRegression
ml=LinearRegression()
ml.fit(x_train,y_train)
```

# predict the test set result

```python
y_pred=ml.predict(x_test)
print(y_pred)
```
