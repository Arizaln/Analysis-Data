```python
# mempersiapkan library
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.utils import to_categorical
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Embedding, Activation, Dropout
from tensorflow.keras.layers import Conv1D, MaxPooling1D, GlobalMaxPooling1D
import numpy as np  #digunakan untuk perhitugan numerik
import pandas as pd #import pandas untuk membaca file csv
from sklearn.model_selection import train_test_split #untuk memecah data training dan test
```

```python
df = pd.read_csv('/content/tweetalay.csv')
df.head()
```

MEMBUAT DATASET UNTUK TRAINING NYA

```python
# mengambil bagian text tweet dan menghitung jumlah textnya (variabel x / varibel input)
text = df["Tweet"].tolist()
print(len(text))
```

```python
# membuat variabel y atau target atau sentimen
y = df["sentimen"]
y = to_categorical(y) #y ubah ke kategorical
print(y)
```

```python
# melihat nilai pada setiap jenis data, 0 = netral, 2 = negatif, 1 = positif
df["sentimen"].value_counts()
```

```python
# men-tokenize text twitter
token = Tokenizer()
token.fit_on_texts(text)
```

```python
# membuat index dari kata-kata
token.index_word
```

```python
# menghitung jumlah kata 
vocab = len(token.index_word)+1
vocab
```

```python
# merubah setiap kalimat menjadi bentuk sequential dari bentuk index tokennya contoh nya seperti ini
# aku = index 1, lagi = index 29, galau = index 2699, nih = index 122
x =['aku lagi galau nih'] 
token.texts_to_sequences(x)
```

```python
# membuat seluruh kalimatnya menjadi bentuk index seperti di atas
encode_text = token.texts_to_sequences(text)
print(encode_text)
```

```python
# menormalisasikan kalimat karena jumlah kata tiap kalimat berbeda maka akan kita ubah menjadi lebih fit, kalau lebih akan dipotong, kalau kurang akan dikasih padding/spasi
max_kata = 100 #panjang maximal kalimat adalah 100
X = pad_sequences(encode_text, maxlen = max_kata, padding="post" ) #jika post diganti "pre" maka 0 nya akan di depan
X
```

```python
# memotong data menggunakan fungsi dari sklearn diatas
X_train, X_test, y_train, y_test = train_test_split(X,y, random_state=40, test_size=0.3, stratify=y)
```

```python
# mengubah kedalam bentuk array numpy
X_train = np.asarray(X_train)
X_test = np.asarray(X_test)
y_train = np.asarray(y_train)
y_test = np.asarray(y_test)
```

MEMBUAT MODEL TRAINING NYA

```python
# menentukan vektor size untuk besar size saat proses embedding, jadi sequence huruf yang merupakan hasil konversi dari kata akan kita ubah menjadi word embedding
# Secara sederhana, word embedding adalah proses konversi sebuat teks menjadi angka.
vec_size = 300 
model = Sequential()
# tambahkan 1 layer untuk mengubah sequance index kata menjadi vektor sebanyak Vocab yang diatas, lalu output sebesar vektor_size, lalu inputnya sebanyak maximal 100 kata
model.add(Embedding(vocab, vec_size, input_length= max_kata))
# Menggunakan Convolutional Neural Networks 1 Dimensi sebanyak filter 64 dan 8 unit Convolution kemudian maxpooling
model.add(Conv1D(64,8, activation="relu"))
model.add(MaxPooling1D(2))
# dropout adalah bentuk configurasi agar tidak overfitting.
model.add(Dropout(0.5))

model.add(Dense(32, activation='relu'))
model.add(Dropout(0.5))
model.add(Dense(16, activation='relu'))
model.add(GlobalMaxPooling1D())
# membuat layer klasifikasi akhir
model.add(Dense(3, activation='softmax'))
model.summary()
```

```python
# membuat compile
model.compile(optimizer="adam", loss="categorical_crossentropy", metrics=['accuracy'])
```

```python
# training
model.fit(X_train, y_train, epochs=100, validation_data =(X_test, y_test))
```

```python
# melakukan klasifikasi kalimat
def get_encode(x):
  x = token.texts_to_sequences(x)
  x = pad_sequences(x, maxlen = max_kata, padding="post")
  return x 
```

```python
# mencoba kalimat
x = ['setiap malem aku selalu sendiri']
x = get_encode(x)
```

```python
# mencoba melakukan prediksi
np.argmax(model.predict(x), axis=-1)
```
