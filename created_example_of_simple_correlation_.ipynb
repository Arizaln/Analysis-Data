# data visualisation
 sebuah usaha kita sebagai seorang analyis untuk menyampaikan insight kepada stakeholder berupa diagram, agar lebih mudah di mengerti dan dibaca dibangding kita menyampaikannya dalam bentuk tabel atau excel.

 data visualisation merupakan sebuah subset dari data preprocesing, sama seperti data storytelling 

 
 
 notes : ketika ingin menentukan diagram visualisasi kita harus ingat tujuan bisnis nya itu apa (harus menjawab pertanyaan bisnis tersebut)


 notes: 3d chart itu lebih baik digunakan ketika ada lebih dari 2 sumbu , mislakan jika ingin  representation data dengan 3 sumbu maka lebih baik 3dchart
(jarang digunakan di dunia perdataan), biasa orang menggunakan pci di machine learning untuk plot datanya jadi 2d

**apa sih yang bisa dilakukan data visualisasi**

1. comparasion
#perbandingan (menunjukan perbedaan antara dua value)
2. distribution
#show occurence of data( seberapa sering data ini muncul),intinya penyebaran data nya tu lebih dapat dianalisis.
3. relationship
#kita pakai data visualisasi untuk menjelaskan hubungan antar variabel bisa dua variabel atau lebih 
kolrelalsi dalam analisis data penting banget, ada 3 macam yaitu positif negatif (klo satu naik satunya lgi turun) dan tidak ada korelasi 
4. composition.
untuk menunjukan porsi dari satu atau lebih variabel dengan sebuah presentase

```python
# lets code
# basic plot
import matplotlib.pyplot as plt
```

```python
# amount of ice cream sold

x=[10,40,50,80]

# temprature in celcius

y=[25,30,35,41]


x2=[5,10,30,80]
y2=[15,15,18,30]


plt.ylabel('temprature in celcius')#untuk memberikan axis label pada sumbu y
plt.xlabel('amount of ice cream sold')#untuk memberikan axis label pada sumbu x
plt.title('sales of ice cream based on temprature ') #untuk memberikan title pada chartnya
plt.plot(x,y,color='red',label='ice cream 1') #untuk memanggil  ,# untuk merubah warna
plt.plot(x2,y2,color='blue',label='ice cream 2') 
plt.legend()#untuk menandai perbandingan antara variabel 1 dan 2
```

```python
bulan=['januari','februari','maret','april','mei','juni','juli','agustus','september','oktober','november','desember']
jumlah=[55,45,60,41,45,60,80,100,70,65,55,85]

bulan2=['januari','februari','maret','april','mei','juni','juli','agustus','september','oktober','november','desember']
jumlah2=[60,55,70,65,45,40,50,85,90,95,85,70]

plt.ylabel('Jumlah pelanggan')
plt.xlabel('Bulan')
plt.title('pertumbuhan pelanggan dalam 12 bulan terakhir di bank maju')
plt.rcParams["figure.figsize"] = (20,10)
plt.plot(bulan,jumlah,color='red',label='jakarta selatan',marker='o') 
plt.plot(bulan2,jumlah2,color='blue',label='depok',marker='o') 
plt.grid(True)
plt.legend()
```

```python
data={
    'jumlah kunjungan':[ 120,130,140,150,170,180],
    'jumlah pelanggan':[ 100,110,120,130,140,160]
    }


```

```python
plt.title('jumlah kunjungan vs jumlah pelanggan') 
plt.ylabel('jumlah kunjungan')
plt.xlabel('jumlah pelanggan')
plt.scatter(df['jumlah pelanggan'],df['jumlah kunjungan'])
plt.grid(True)
```
